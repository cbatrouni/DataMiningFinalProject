```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
import category_encoders as ce
from sklearn.model_selection import train_test_split

%matplotlib inline
```

```python
df=pd.read_csv("/kaggle/input/income/adult.csv")

#check the head
df.head()
```

```python
col_names = ['age', 'workclass', 'fnlwgt', 'education', 'education_num', 'marital_status', 'occupation', 'relationship',
             'race', 'sex', 'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'income']

df.columns = col_names

#checks column names that we just set
df.columns
```

```python
X = df.drop(['income'], axis=1)

y = df['income']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
encoder = ce.OneHotEncoder(cols=['workclass', 'education', 'marital_status', 'occupation', 'relationship', 
                                 'race', 'sex', 'native_country'])

X_train = encoder.fit_transform(X_train)

X_test = encoder.transform(X_test)
```

```python
#
naive_bayes = GaussianNB()
# fit the model
naive_bayes.fit(X_train, y_train)
```

```python
y_pred = naive_bayes.predict(X_test)

y_pred
```

```python
sns.pairplot(df,hue='race', height=2.5)
```

```python
print('Accuracy Score: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))

print('Testset score: {:.4f}'.format(naive_bayes.score(X_test, y_test)))
print('Trainingset score: {:.4f}'.format(naive_bayes.score(X_train, y_train)))
```
